```python
import os
try:
    from dotenv import load_dotenv
    load_dotenv()
except ImportError:
    import subprocess, sys
    subprocess.check_call([sys.executable, "-m", "pip", "install", "python-dotenv", "-q"])
    from dotenv import load_dotenv
    load_dotenv()

import os, sys, subprocess, warnings, gc, math, random, shutil, json, torch, numpy as np, pandas as pd, matplotlib.pyplot as plt, matplotlib as mpl, seaborn as sns, spacy
from itertools import product; from typing import Dict, Any, List; from google.colab import drive
from mpl_toolkits.mplot3d import Axes3D; from mpl_toolkits.mplot3d.art3d import Line3DCollection
from scipy.linalg import svd as scipy_svd; from scipy.stats import kurtosis, gaussian_kde
from scipy.optimize import linear_sum_assignment; from tqdm.notebook import tqdm
from transformers import AutoModelForCausalLM, AutoTokenizer, TrainingArguments, Trainer, DataCollatorForLanguageModeling
from datasets import load_dataset, Dataset; from peft import LoraConfig, get_peft_model
# dataset in json
X='x.json'
def install_packages_if_needed():
    try: import peft, spacy, xlsxwriter, statsmodels
    except ImportError: subprocess.check_call([sys.executable, "-m", "pip", "install", "transformers", "datasets", "peft", "torch", "accelerate", "statsmodels", "pandas", "spacy", "matplotlib", "scipy", "tqdm", "xlsxwriter", "-q"]); subprocess.call([sys.executable, "-m", "spacy", "download", "en_core_web_sm", "-q"])

install_packages_if_needed(); nlp = spacy.load("en_core_web_sm") if 'spacy' in sys.modules else lambda x: x
try: drive.mount('/content/drive', force_remount=True)
except: pass
warnings.filterwarnings('ignore'); plt.rcParams.update({"font.family": "serif", "font.serif": ["cmr10", "DejaVu Serif", "Bitstream Vera Serif"], "mathtext.fontset": "cm", "axes.labelsize": 14, "font.size": 12, "legend.fontsize": 12, "xtick.labelsize": 12, "ytick.labelsize": 12, "figure.figsize": [10, 8], "image.cmap": "jet", "axes.formatter.use_mathtext": True})
BASE_DRIVE_PATH = '/content/drive/My Drive/University_MATH'; DATASET_FILENAME = X; DATASET_FULL_PATH = os.path.join(BASE_DRIVE_PATH, DATASET_FILENAME)

QUICK_CONFIG = {"run_config": {"experiment_name": "Exp_Quick", "num_replicas": 1, "device": "cuda"}, "data_config": {"dataset_path": DATASET_FULL_PATH, "linguistic_sample_size": 100, "finetune_budget": 10, "max_seq_len": 64}, "model_config": {"base_model_id": "openai-community/gpt2", "lora_config": {"r": 4, "lora_alpha": 8, "target_modules": ["c_attn"]}}, "training_config": {"max_steps": 5, "learning_rate": 1e-4, "per_device_train_batch_size": 4, "logging_steps": 2}, "rlct_config": {"max_rlct_estimation_steps": 5, "sgld_lr": 1e-4}, "lot_config": {"use_alternating_source": True, "dolly_lot_sample_size": 10, "n_tokens": 10, "num_trajectories": 10}}
INTERMEDIATE_CONFIG = {"run_config": {"experiment_name": "Exp_PhD_2_Intermediate", "num_replicas": 1, "device": "cuda"}, "data_config": {"dataset_path": DATASET_FULL_PATH, "linguistic_sample_size": 1000, "finetune_budget": 50, "max_seq_len": 128}, "model_config": {"base_model_id": "openai-community/gpt2", "lora_config": {"r": 8, "lora_alpha": 16, "target_modules": ["c_attn"]}}, "training_config": {"max_steps": 20, "learning_rate": 5e-5, "per_device_train_batch_size": 4, "logging_steps": 5}, "rlct_config": {"max_rlct_estimation_steps": 15, "sgld_lr": 1e-4}, "lot_config": {"use_alternating_source": True, "dolly_lot_sample_size": 30, "n_tokens": 20, "num_trajectories": 50}}
DEEP_CONFIG = {"run_config": {"experiment_name": "Exp_PhD_3_Deep", "num_replicas": 2, "device": "cuda"}, "data_config": {"dataset_path": DATASET_FULL_PATH, "linguistic_sample_size": 5000, "finetune_budget": 150, "max_seq_len": 128}, "model_config": {"base_model_id": "openai-community/gpt2", "lora_config": {"r": 16, "lora_alpha": 32, "target_modules": ["c_attn"]}}, "training_config": {"max_steps": 50, "learning_rate": 5e-5, "per_device_train_batch_size": 8, "logging_steps": 10}, "rlct_config": {"max_rlct_estimation_steps": 50, "sgld_lr": 5e-5}, "lot_config": {"use_alternating_source": True, "dolly_lot_sample_size": 100, "n_tokens": 30, "num_trajectories": 100}}
ACTIVE_CONFIG = QUICK_CONFIG

class ExperimentDocumentationGenerator:
    def __init__(self, output_dir, config): self.base_dir, self.doc_dir, self.config = output_dir, os.path.join(output_dir, "Mathematical_Foundations_and_Analysis"), config; os.makedirs(self.doc_dir, exist_ok=True)
    def generate_all(self): self._write_math_foundations(); self._write_graph_analysis(); self._write_conclusions(); print(f"   -> Documentation saved in: {self.doc_dir}")
    def _write_math_foundations(self):
        with open(os.path.join(self.doc_dir, "01_Mathematical_Foundations.txt"), "w") as f: f.write(r"soon...")
    def _write_graph_analysis(self):
        with open(os.path.join(self.doc_dir, "02_Graphical_Analysis.txt"), "w") as f: f.write(r"soon")
    def _write_conclusions(self):
        with open(os.path.join(self.doc_dir, "03_Conclusions_and_Implications.txt"), "w") as f: f.write(r"soon...")

class SingularLearningEstimator:
    class SGLD(torch.optim.Optimizer):
        def __init__(self, params, lr=1e-3, noise_level=1., temperature=1., num_samples=1, batch_size=1): super().__init__(params, dict(lr=lr, noise_level=noise_level, temperature=temperature, num_samples=num_samples, batch_size=batch_size))
        def step(self, closure=None):
            for group in self.param_groups:
                for p in group['params']:
                    if p.grad is not None:
                        with torch.no_grad(): p.add_(p.grad * (group["num_samples"] / group["batch_size"]), alpha=-0.5 * group['lr'] / group['temperature']); p.add_(torch.normal(mean=0., std=group['noise_level'], size=p.grad.size(), device=p.grad.device), alpha=math.sqrt(group['lr']))
    @staticmethod
    def estimate_rlct(model, tokenizer, dataset, config):
        print("   -> RLCT Phase: WBIC Estimation (Watanabe 2013)..."); n = len(dataset); beta1, beta2 = 1.0/np.log(n), 1.5/np.log(n); optimizer = SingularLearningEstimator.SGLD(model.parameters(), lr=config['sgld_lr'], num_samples=n, batch_size=1, temperature=1.0/beta1); energies = []; model.eval()
        for i in tqdm(range(min(n, config['max_rlct_estimation_steps'])), desc="WBIC Sampling", leave=False):
            inputs = tokenizer(f"{dataset[i%n]['instruction']}\n{dataset[i%n]['response']}", truncation=True, max_length=config["max_seq_len"], return_tensors="pt").to(config['device']); model.train(); optimizer.zero_grad(); loss = model(**inputs, labels=inputs.input_ids).loss; loss.backward(); optimizer.step(); model.eval(); energies.append(loss.item() * n)
        if not energies: return 0.0, model, []
        V = np.array(energies); E_b1 = np.mean(V); log_w = -(beta2 - beta1) * V; norm_w = np.exp(log_w - np.max(log_w)); E_b2 = np.sum((norm_w/np.sum(norm_w)) * V); return (E_b1 - E_b2) / ((1.0/beta1) - (1.0/beta2)), model, [e/n for e in energies]

class ScientificLoTSuite:
    def __init__(self, model, tokenizer, output_dir): self.model, self.tokenizer, self.output_dir = model, tokenizer, output_dir; os.makedirs(output_dir, exist_ok=True); self.model.eval()
    def generate_trajectories(self, text, lot_config):
        print(f"   -> Generating {lot_config['num_trajectories']} LoT trajectories..."); tokens = self.tokenizer(text, return_tensors='pt', truncation=True, max_length=1024).input_ids[0]; sentences = [tokens[i:i+lot_config['n_tokens']] for i in range(0,len(tokens)-lot_config['n_tokens']+1, lot_config['n_tokens'])][:lot_config['num_trajectories']]
        if not sentences: return None
        trajectories = []
        with torch.no_grad():
            for s in tqdm(sentences, desc="Tracing", leave=False): trajectories.append(np.stack([h[0,-1,:].float().cpu().numpy() for h in self.model(s.unsqueeze(0).to(self.model.device), output_hidden_states=True).hidden_states], axis=1))
        return np.stack(trajectories, axis=2)[:, 1:, :] if trajectories else None
    def _plot_ribbon_trajectories(self, traj):
        h, l, s = traj.shape; flat = traj.reshape(h, -1).astype(np.float32)
        try: U, S, V = scipy_svd(flat, full_matrices=False)
        except: return
        proj = (np.diag(S[:3]) @ V[:3, :]).reshape(3, l, s); fig = plt.figure(figsize=(10, 8)); ax = fig.add_subplot(111, projection='3d'); cmap = plt.get_cmap('jet', l)
        for i in range(min(s, 50)): points = np.array([proj[0,:,i], proj[1,:,i], proj[2,:,i]]).T.reshape(-1, 1, 3); ax.add_collection(Line3DCollection(np.concatenate([points[:-1], points[1:]], axis=1), cmap=cmap, norm=plt.Normalize(0, l), alpha=0.4, linewidth=0.8))
        ax.set_xlim(proj[0].min(), proj[0].max()); ax.set_ylim(proj[1].min(), proj[1].max()); ax.set_zlim(proj[2].min(), proj[2].max()); ax.set_xlabel(r'$\mathbf{u}_1$', labelpad=10); ax.set_ylabel(r'$\mathbf{u}_2$', labelpad=10); ax.set_zlabel(r'$\mathbf{u}_3$', labelpad=10); ax.set_title(r'Lines of Thought Manifold (PCA)', y=1.02); ax.grid(False); fig.colorbar(plt.cm.ScalarMappable(cmap=cmap, norm=plt.Normalize(0, l)), ax=ax, shrink=0.6, pad=0.1).set_label(r'Layer Depth $t$', rotation=270, labelpad=20); plt.savefig(f"{self.output_dir}/lot_manifold_pca_projection.png", dpi=300, bbox_inches='tight'); plt.close()
    def _plot_coordinate_manifold(self, traj):
        h, l, s = traj.shape; vars_dim = np.var(traj.reshape(h, -1), axis=1); top = np.argsort(vars_dim)[-3:]; fig = plt.figure(figsize=(10, 8)); ax = fig.add_subplot(111, projection='3d'); cmap = plt.get_cmap('viridis', l)
        for t in range(l): ax.scatter(traj[top[0], t, ::2], traj[top[1], t, ::2], traj[top[2], t, ::2], color=cmap(t), s=3, alpha=0.6)
        ax.set_xlabel(rf'$\mathbf{{e}}_{{{top[0]}}}$'); ax.set_ylabel(rf'$\mathbf{{e}}_{{{top[1]}}}$'); ax.set_zlabel(rf'$\mathbf{{e}}_{{{top[2]}}}$'); ax.set_title(r'Manifold in Coordinate Basis', y=1.02); fig.colorbar(plt.cm.ScalarMappable(cmap=cmap, norm=mpl.colors.Normalize(vmin=0, vmax=l)), ax=ax, shrink=0.6, pad=0.1).set_label(r'Layer Depth $t$', rotation=270, labelpad=20); plt.savefig(f"{self.output_dir}/lot_coordinate_basis_projection.png", dpi=300, bbox_inches='tight'); plt.close()
    def _plot_sv_angles(self, traj):
        n = traj.shape[1]; us = [scipy_svd(traj[:,t,:].astype(np.float32), full_matrices=False)[0] for t in range(n)]; fig, axes = plt.subplots(2, 2, figsize=(12, 10)); fig.suptitle(r'Angle $\arccos(\mathbf{u}_i^{(t_1)} \cdot \mathbf{u}_i^{(t_2)})$', y=0.95, fontsize=16)
        for i in range(4):
            mat = np.array([[np.degrees(np.arccos(np.clip(np.abs(np.dot(us[t1][:, i], us[t2][:, i])), -1.0, 1.0))) for t2 in range(n)] for t1 in range(n)])
            im = axes[i//2, i%2].imshow(mat, cmap='viridis', origin='lower', vmin=0, vmax=90); axes[i//2, i%2].set_title(rf'Singular Vector $u_{i+1}$'); fig.colorbar(im, ax=axes[i//2, i%2], fraction=0.046, pad=0.04).set_label(r'$\theta$ (deg)', rotation=270, labelpad=15)
        plt.tight_layout(rect=[0, 0.03, 1, 0.93]); plt.savefig(f"{self.output_dir}/svd_singular_vector_alignment_angles.png", dpi=300); plt.close()
    def _plot_singular_values(self, traj):
        n = traj.shape[1]; sigs = [scipy_svd(traj[:,t,:].astype(np.float32), full_matrices=False, compute_uv=False) for t in range(n)]; fig, ax = plt.subplots(figsize=(10, 8)); cmap = plt.get_cmap('jet', n)
        for t in range(n): ax.loglog(range(1, len(sigs[t])+1), sigs[t], color=cmap(t), alpha=0.9, linewidth=1.5)
        ax.set_title(r'Singular Value Spectrum'); ax.set_xlabel(r'Rank $k$'); ax.set_ylabel(r'$\sigma_k$'); ax.grid(True, which="both", ls="-", alpha=0.2); fig.colorbar(plt.cm.ScalarMappable(cmap=cmap, norm=mpl.colors.Normalize(vmin=0, vmax=n)), ax=ax).set_label(r'Layer $t$', rotation=270, labelpad=20); plt.savefig(f"{self.output_dir}/svd_singular_value_spectrum_decay.png", dpi=300, bbox_inches='tight'); plt.close()
    def _plot_kl_vs_k(self, traj):
        final = torch.tensor(traj[:, -1, :].T).to(self.model.device).float(); head = self.model.get_output_embeddings().weight.T.float(); kl = []
        with torch.no_grad():
            true_p = torch.softmax(final @ head, dim=-1); U = torch.svd(final.T)[0]; max_rank = min(U.shape[1], final.shape[1], 128); K_dims = np.unique(np.geomspace(1, max(2, max_rank-1), 10, dtype=int))
            for k in K_dims: kl.append(torch.nn.functional.kl_div(torch.log_softmax(((U[:, :k] @ U[:, :k].T) @ final.T).T @ head, dim=-1), true_p, reduction='batchmean').item())
        fig, ax = plt.subplots(figsize=(8, 6)); ax.plot(K_dims, kl, 'k.-', linewidth=1.5); ax.set_title(r'$\langle D_{KL}(\mathbf{p}_K^\nu || \mathbf{p}^\nu) \rangle$ vs Rank $K$'); plt.savefig(f"{self.output_dir}/kl_divergence_information_loss_vs_rank.png", dpi=300); plt.close()
    def _plot_delta_heatmaps(self, traj):
        n = traj.shape[1]; mat_m, mat_v, mat_k = np.full((n,n), np.nan), np.full((n,n), np.nan), np.full((n,n), np.nan)
        for t in range(n):
            for tau in range(1, n - t): d = traj[:, t + tau, :] - traj[:, t, :]; mat_m[t+tau, t] = np.mean(d)/(np.std(d)+1e-9); mat_v[t+tau, t] = np.log(np.var(d)+1e-9); mat_k[t+tau, t] = np.mean(kurtosis(d, axis=1))
        fig, axes = plt.subplots(1, 3, figsize=(18, 5)); labs = [r'$\langle |\mu/\sigma| \rangle$', r'$\langle \log(\sigma^2) \rangle$', r'$\langle |\kappa| \rangle$']; mats = [mat_m, mat_v, mat_k]
        for i in range(3): im = axes[i].imshow(mats[i], cmap='jet', aspect='auto', origin='upper'); axes[i].set_title(labs[i]); fig.colorbar(im, ax=axes[i])
        plt.savefig(f"{self.output_dir}/drift_diffusion_statistics_heatmaps.png", dpi=300, bbox_inches='tight'); plt.close()
    def _plot_kde_residuals(self, traj):
        h, l, s = traj.shape; fig, ax = plt.subplots(figsize=(8, 6)); d8, d10 = traj[:, 8, :] - traj[:, 0, :], traj[:, 10, :] - traj[:, 0, :]
        sns.kdeplot((d8.flatten()-np.mean(d8))/(np.std(d8)+1e-9), ax=ax, label=r'$t+\tau=8$', color='tab:blue'); sns.kdeplot((d10.flatten()-np.mean(d10))/(np.std(d10)+1e-9), ax=ax, label=r'$t+\tau=10$', color='tab:orange')
        x = np.linspace(-5, 5, 100); ax.plot(x, 1/np.sqrt(2*np.pi)*np.exp(-0.5*x**2), 'k--', label='Gaussian'); ax.set_yscale('log'); ax.legend(); plt.savefig(f"{self.output_dir}/residual_distribution_normality_check.png", dpi=300); plt.close()
        try:
            flat = traj.reshape(h, -1).astype(np.float32); U, S, V = scipy_svd(flat, full_matrices=False); proj = (np.diag(S[:2]) @ V[:2, :]).reshape(2, l, s); fig, axes = plt.subplots(2, 5, figsize=(20, 8)); shifts = [6, 7, 8, 9, 10]
            for i, sh in enumerate(shifts):
                if sh < l:
                    for row, t_idx in enumerate([sh, sh-2]):
                        u1, u2 = proj[0, t_idx, :], proj[1, t_idx, :]; k = gaussian_kde([u1, u2]); xi, yi = np.mgrid[u1.min():u1.max():20j, u2.min():u2.max():20j]; axes[row, i].contour(xi, yi, k(np.vstack([xi.flatten(), yi.flatten()])).reshape(xi.shape), cmap='jet')
            plt.savefig(f"{self.output_dir}/probability_density_evolution_contours.png", dpi=300); plt.close()
        except: pass
    def run_full_analysis(self, lot_config, dolly_dataset):
        print(f"   -> Executing LoT Analysis in: {os.path.basename(self.output_dir)}")
        if lot_config.get('use_alternating_source', True): sample = dolly_dataset.shuffle().select(range(min(len(dolly_dataset), 100))); text = "\n\n".join(list(sample['instruction']) + [r for r in sample['response'] if r])
        else: text = " ".join(["thought"] * 500)
        traj = self.generate_trajectories(text, lot_config)
        if traj is not None: [getattr(self, m)(traj) for m in dir(self) if m.startswith('_plot_')]; print("   -> Scientific Charts Generated.")

def plot_training_history(history, run_id, output_dir):
    if history:
        df = pd.DataFrame(history).dropna(subset=['loss']); plt.figure(figsize=(10, 6)); plt.plot(df['step'], df['loss'], marker='o', label='Training Loss'); plt.title(rf"Training Loss Run {run_id}"); plt.yscale('log'); plt.grid(True, alpha=0.3); plt.savefig(os.path.join(output_dir, "training_history_loss.png")); plt.close()

def run_linguistic_anova_experiment(master_config: Dict[str, Any]):
    run_cfg, data_cfg = master_config['run_config'], master_config['data_config']; output_dir = os.path.join("/content", run_cfg['experiment_name']); os.makedirs(output_dir, exist_ok=True); doc_gen = ExperimentDocumentationGenerator(output_dir, master_config)
    try: full_ds = load_dataset('json', data_files=data_cfg['dataset_path'], split='train')
    except: full_ds = Dataset.from_pandas(pd.DataFrame({"instruction": ["Explain AI", "Solve 1+1"]*50, "response": ["AI is code", "2"]*50}))
    subset = full_ds.shuffle(seed=42).select(range(min(data_cfg['linguistic_sample_size'], len(full_ds)))); df = pd.DataFrame([{'syntax': len(t.split()), 'lexicon': len(set(t.split()))/len(t.split()) if t.split() else 0, **subset[i]} for i, t in enumerate([str(t) for t in subset['instruction']])]); b = data_cfg['finetune_budget']; pools = {"Simple Syntax": Dataset.from_pandas(df.sort_values('syntax').head(b)), "Complex Syntax": Dataset.from_pandas(df.sort_values('syntax', ascending=False).head(b)), "Rich Lexicon": Dataset.from_pandas(df.sort_values('lexicon', ascending=False).head(b))}; results, all_rlct = [], []; tokenizer = AutoTokenizer.from_pretrained(master_config['model_config']['base_model_id']); tokenizer.pad_token = tokenizer.eos_token or tokenizer.pad_token; run_counter = 0
    for replica in range(run_cfg['num_replicas']):
        for treat, method in product(pools.keys(), ["Standard LoRA", "Two-Phase LoRA + RLCT"]):
            run_counter += 1; run_id = f"{run_counter}_Rep{replica}_{treat.replace(' ','')}_{method.split()[0]}"; print(f"\nRun {run_id} | {treat} | {method}"); run_dir = os.path.join(output_dir, run_id); os.makedirs(run_dir, exist_ok=True); peft_model = get_peft_model(AutoModelForCausalLM.from_pretrained(master_config['model_config']['base_model_id'], dtype=torch.float32).to(run_cfg['device']), LoraConfig(task_type="CAUSAL_LM", **master_config['model_config']['lora_config'])); rlct_val = np.nan
            if "RLCT" in method: rlct_val, peft_model, losses = SingularLearningEstimator.estimate_rlct(peft_model, tokenizer, pools[treat], {**master_config['rlct_config'], "max_seq_len": data_cfg["max_seq_len"], "device": run_cfg["device"]}); all_rlct.append(pd.DataFrame({'run': run_id, 'loss': losses}))
            trainer = Trainer(model=peft_model, args=TrainingArguments(output_dir=f"{run_dir}/ckpt", report_to="none", save_strategy="no", remove_unused_columns=False, **master_config['training_config']), train_dataset=pools[treat], data_collator=lambda b: {'input_ids': tokenizer([f"Instruction:\n{x['instruction']}\n\nResponse:\n{x['response']}" for x in b], truncation=True, padding="max_length", max_length=data_cfg["max_seq_len"], return_tensors="pt")["input_ids"], 'labels': tokenizer([f"Instruction:\n{x['instruction']}\n\nResponse:\n{x['response']}" for x in b], truncation=True, padding="max_length", max_length=data_cfg["max_seq_len"], return_tensors="pt")["input_ids"]}); trainer.train(); plot_training_history(trainer.state.log_history, run_id, run_dir); peft_model.eval(); ScientificLoTSuite(peft_model, tokenizer, os.path.join(run_dir, "LoT_Analysis")).run_full_analysis(master_config['lot_config'], full_ds)
            results.append({"Run": run_id, "Treatment": treat, "Method": method, "RLCT": rlct_val, "PPL": math.exp(np.mean([peft_model(**tokenizer(f"{x['instruction']}\n{x['response']}", return_tensors='pt', truncation=True, max_length=128).to(run_cfg['device']), labels=tokenizer(f"{x['instruction']}\n{x['response']}", return_tensors='pt', truncation=True, max_length=128).to(run_cfg['device']).input_ids).loss.item() for x in full_ds.shuffle(seed=123).select(range(min(20, len(full_ds))))]))}); del peft_model, trainer; gc.collect(); torch.cuda.empty_cache()
    with pd.ExcelWriter(os.path.join(output_dir, "FINAL_RESULTS.xlsx"), engine='xlsxwriter') as w: pd.DataFrame(results).to_excel(w, sheet_name='Main', index=False); (pd.concat(all_rlct).to_excel(w, sheet_name='RLCT_Loss', index=False) if all_rlct else None)
    doc_gen.generate_all(); dest = os.path.join(BASE_DRIVE_PATH, f"Results_{run_cfg['experiment_name']}"); shutil.copytree(output_dir, dest, dirs_exist_ok=True) if os.path.exists(BASE_DRIVE_PATH) else print(f"\nSaved locally: {output_dir}")
print(f"STARTING EXPERIMENT: {ACTIVE_CONFIG['run_config']['experiment_name']}"); run_linguistic_anova_experiment(ACTIVE_CONFIG)
```
